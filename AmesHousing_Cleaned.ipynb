# Ames Housing Dataset Cleaning

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 120)
```

```python
# Load the raw dataset. Keep AmesHousing.csv in the same folder as this notebook.
df = pd.read_csv("AmesHousing.csv")
raw_shape = df.shape
df.head()
```

```python
raw_shape, df.info()
```

```python
# Keep a copy of the original data for checking changes later.
original_df = df.copy()
```

```python
# Remove exact repeated records, if any.
rows_before = len(df)
df = df.drop_duplicates().reset_index(drop=True)
rows_removed = rows_before - len(df)
rows_removed
```

```python
# Drop columns that are identifiers or almost constant in this dataset.
cols_to_drop = [col for col in ["Order", "PID", "Utilities", "Street"] if col in df.columns]
df = df.drop(columns=cols_to_drop)
cols_to_drop
```

```python
# Standardize text values by removing accidental spaces around category names.
text_columns = df.select_dtypes(include="object").columns
for col in text_columns:
    df[col] = df[col].str.strip()
```

```python
# Fix a clear typo found in the garage year column.
if "Garage Yr Blt" in df.columns:
    df.loc[df["Garage Yr Blt"] == 2207, "Garage Yr Blt"] = 2007
    df["Garage Yr Blt"] = df["Garage Yr Blt"].where(df["Garage Yr Blt"].le(2026), np.nan)
```

```python
# Missing categorical values where blank usually means the house does not have that feature.
none_like_columns = [
    "Alley", "Bsmt Qual", "Bsmt Cond", "Bsmt Exposure", "BsmtFin Type 1",
    "BsmtFin Type 2", "Fireplace Qu", "Garage Type", "Garage Finish",
    "Garage Qual", "Garage Cond", "Pool QC", "Fence", "Misc Feature",
    "Mas Vnr Type"
]
for col in none_like_columns:
    if col in df.columns:
        df[col] = df[col].fillna("None")
```

```python
# Missing numeric values where blank means area/count is zero.
zero_value_columns = [
    "Mas Vnr Area", "BsmtFin SF 1", "BsmtFin SF 2", "Bsmt Unf SF",
    "Total Bsmt SF", "Bsmt Full Bath", "Bsmt Half Bath",
    "Garage Cars", "Garage Area"
]
for col in zero_value_columns:
    if col in df.columns:
        df[col] = df[col].fillna(0)
```

```python
# Fill Lot Frontage using neighborhood median first, then global median for leftovers.
if {"Neighborhood", "Lot Frontage"}.issubset(df.columns):
    frontage_by_area = df.groupby("Neighborhood")["Lot Frontage"].transform("median")
    df["Lot Frontage"] = df["Lot Frontage"].fillna(frontage_by_area)

if "Lot Frontage" in df.columns:
    df["Lot Frontage"] = df["Lot Frontage"].fillna(df["Lot Frontage"].median())
```

```python
# Use house build year when garage build year is missing.
if {"Garage Yr Blt", "Year Built"}.issubset(df.columns):
    df["Garage Yr Blt"] = df["Garage Yr Blt"].fillna(df["Year Built"])
```

```python
# Fill remaining categorical missing values with the most frequent value.
remaining_cat = df.select_dtypes(include="object").columns
for col in remaining_cat:
    if df[col].isna().any():
        df[col] = df[col].fillna(df[col].mode(dropna=True)[0])
```

```python
# Fill remaining numeric missing values with median values.
remaining_num = df.select_dtypes(include=np.number).columns
for col in remaining_num:
    if df[col].isna().any():
        df[col] = df[col].fillna(df[col].median())
```

```python
# Convert columns that are codes/time labels, not continuous measurements.
for col in ["MS SubClass", "Mo Sold"]:
    if col in df.columns:
        df[col] = df[col].astype("category")
```

```python
# Encode ordered quality columns using meaningful ranks.
quality_rank = {"None": 0, "Po": 1, "Fa": 2, "TA": 3, "Gd": 4, "Ex": 5}
quality_columns = [
    "Exter Qual", "Exter Cond", "Bsmt Qual", "Bsmt Cond",
    "Heating QC", "Kitchen Qual", "Fireplace Qu",
    "Garage Qual", "Garage Cond", "Pool QC"
]
for col in quality_columns:
    if col in df.columns:
        df[col] = df[col].replace(quality_rank)
        df[col] = df[col].infer_objects(copy=False).astype(int)
```

```python
# Encode other ordered text features.
ordered_maps = {
    "Functional": {"Sal": 1, "Sev": 2, "Maj2": 3, "Maj1": 4, "Mod": 5, "Min2": 6, "Min1": 7, "Typ": 8},
    "Bsmt Exposure": {"None": 0, "No": 1, "Mn": 2, "Av": 3, "Gd": 4},
    "BsmtFin Type 1": {"None": 0, "Unf": 1, "LwQ": 2, "Rec": 3, "BLQ": 4, "ALQ": 5, "GLQ": 6},
    "BsmtFin Type 2": {"None": 0, "Unf": 1, "LwQ": 2, "Rec": 3, "BLQ": 4, "ALQ": 5, "GLQ": 6},
    "Garage Finish": {"None": 0, "Unf": 1, "RFn": 2, "Fin": 3},
    "Paved Drive": {"N": 0, "P": 1, "Y": 2},
    "Land Slope": {"Sev": 1, "Mod": 2, "Gtl": 3},
    "Lot Shape": {"IR3": 1, "IR2": 2, "IR1": 3, "Reg": 4},
}

for col, mapping in ordered_maps.items():
    if col in df.columns:
        df[col] = df[col].map(mapping).fillna(df[col].map(mapping).median()).astype(int)
```

```python
# Convert binary yes/no features into 1/0.
binary_maps = {
    "Central Air": {"N": 0, "Y": 1},
}
for col, mapping in binary_maps.items():
    if col in df.columns:
        df[col] = df[col].map(mapping).astype(int)
```

```python
# Cap extreme numeric outliers using the IQR method. SalePrice is kept unchanged as target variable.
def cap_iqr(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return series.clip(lower=lower, upper=upper)

numeric_for_capping = df.select_dtypes(include=np.number).columns.drop("SalePrice", errors="ignore")
for col in numeric_for_capping:
    if df[col].nunique() > 15:
        df[col] = cap_iqr(df[col])
```

```python
# Add simple age-related features useful for later modelling.
if {"Yr Sold", "Year Built"}.issubset(df.columns):
    df["House Age"] = (df["Yr Sold"] - df["Year Built"]).clip(lower=0)

if {"Yr Sold", "Year Remod/Add"}.issubset(df.columns):
    df["Years Since Remodel"] = (df["Yr Sold"] - df["Year Remod/Add"]).clip(lower=0)
```

```python
# Final missing-value check.
missing_after = df.isna().sum().sort_values(ascending=False)
missing_after[missing_after > 0]
```
