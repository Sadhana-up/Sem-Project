# FIX Dtypes

```python
df.dtypes
```

```python
df["MS SubClass"] = df["MS SubClass"].astype(str) ## to avoid category
```

```python
int_year_cols = ["Year Built", "Year Remod/Add", "Yr Sold", "Mo Sold"]
df[int_year_cols] = df[int_year_cols].astype(int)
```

### Handle Missing Values In Categorical columns

```python
categorical_none_cols = ["Pool QC", "Misc Feature", "Alley", "Fence","Fireplace Qu","Mas Vnr Type","Garage Type", "Garage Finish", "Garage Qual", "Garage Cond",
"Bsmt Qual", "Bsmt Cond", "Bsmt Exposure",
    "BsmtFin Type 1", "BsmtFin Type 2",
] ## for categorical filled with none
```

```python
df[categorical_none_cols] = df[categorical_none_cols].fillna("None")
```

### Handle Missing Values In Numerical columns

```python
zero_fill_cols = [
    "BsmtFin SF 1", "BsmtFin SF 2", "Bsmt Unf SF", "Total Bsmt SF",
    "Bsmt Full Bath", "Bsmt Half Bath",
    "Garage Cars", "Garage Area",
]
df[zero_fill_cols] = df[zero_fill_cols].fillna(0)
```

```python
df["Mas Vnr Area"] = df["Mas Vnr Area"].fillna(0)
```

```python
df["Garage Yr Blt"] = df["Garage Yr Blt"].fillna(df["Year Built"]) ## garage ta house build hune year mei built va hunxa so 
```

```python
df["Lot Frontage"].describe()
```

```python
## checking skewness to whether use mean or median
```

```python
df["Lot Frontage"].skew()
```

```python
# since it is right skewed there are outliers so use median
```

```python
nbhd_median = df.groupby("Neighborhood")["Lot Frontage"].median().sort_values()
```

```python
nbhd_median
```

```python
global_median = df["Lot Frontage"].median()  
df["Lot Frontage"] = df["Lot Frontage"].fillna(global_median)
```

```python
df.isnull().sum()
```

```python
df['Electrical'].isnull().sum()
```

```python
df["Electrical"] = df["Electrical"].fillna(df["Electrical"].mode()[0])
```

```python
df['Electrical'].isnull().sum()
```

### Removing Whitespaces in Strings

```python
str_cols = df.select_dtypes(include=["object"]).columns ## includes the col which are object 
df[str_cols] = df[str_cols].apply(lambda col: col.str.strip())
```

```python
cat_cols = df.select_dtypes(include=["object"]).columns
df[cat_cols] = df[cat_cols].astype("category") ## change it back to categorical data type 
```
