```python
# removing unnecessary columns in the dataset as these doesn;t help in predicting the house price at all
df.drop(['Order', 'PID','Utilities'], axis=1, inplace=True)
```

### filling the null values 

```python
# filling some columns with none since they are blank not because iof missing values but becuz the house genuienly don't have the feature 
# and they are categorical data too and zero for numeric
# for categorical fill with 'None' string
cat_cols = ['Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu', 'Mas Vnr Type',
            'Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond',
            'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[cat_cols] = df[cat_cols].fillna('None')

# for numeric we fill with  zero
num_cols = ['Garage Yr Blt', 'Bsmt Full Bath', 'Bsmt Half Bath',
            'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF']
df[num_cols] = df[num_cols].fillna(0)
```

```python
df['Lot Frontage']=df['Lot Frontage'].fillna(df['Lot Frontage'].median()) #filling with median
```

```python
df['Electrical']=df['Electrical'].fillna(df['Electrical'].mode()[0]) #filling with mode
```

```python
df['Mas Vnr Area'] = df['Mas Vnr Area'].fillna(0)# since the house doesn’t have masonry veneer
```

```python
df['Garage Area'] = df['Garage Area'].fillna(df['Garage Area'].median())
df['Garage Cars'] = df['Garage Cars'].fillna(df['Garage Cars'].median()) # the house has a garage but specs ar not known
```

```python
df.isnull().sum().sort_values(ascending=False)
```

### correcting the datatype as models could get misguided while training if we keep this as it is

```python
df['MS SubClass'] = df['MS SubClass'].astype('category')# since they represent house category nothing of that use
```

```python
df['Mo Sold'] = df['Mo Sold'].astype('category')
```

```python
# we can see many columns like Bsmt Qual ,Kitchen Qual and many more are using ex,gd,ta and more as quality scales whuch are obviously not
# understood by model so we assign the number to this quality scale
qual_map = {'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5}

ord_cols = [
    'Exter Qual','Exter Cond','Bsmt Qual','Bsmt Cond',
    'Heating QC','Kitchen Qual','Fireplace Qu',
    'Garage Qual','Garage Cond','Pool QC'
]

for col in ord_cols:
    df[col] = df[col].fillna('None').replace('None', 0).map(qual_map).fillna(0)
```

```python
# correcting the columns where values are stored as text, but have a clear order since they actually have a hidden order/ranking
# functional
functional_map = {
    'Sal':1, 'Sev':2, 'Maj2':3, 'Maj1':4,
    'Mod':5, 'Min2':6, 'Min1':7, 'Typ':8
}

bsmt_exp_map = {'None':0, 'No':1, 'Mn':2, 'Av':3, 'Gd':4}

bsmt_fin_map = {
    'None':0, 'Unf':1, 'LwQ':2, 'Rec':3,
    'BLQ':4, 'ALQ':5, 'GLQ':6
}

garage_finish_map = {'None':0, 'Unf':1, 'RFn':2, 'Fin':3}



#Paved Drive
drive_map = {'N':0, 'P':1, 'Y':2}

# Land Slope
slope_map = {'Gtl':1, 'Mod':2, 'Sev':3}

# Lot Shape
shape_map = {'IR3':1, 'IR2':2, 'IR1':3, 'Reg':4}
```

```python
# applying everything for the columns 
df['Functional'] = df['Functional'].map(functional_map)

df['Bsmt Exposure'] = df['Bsmt Exposure'].map(bsmt_exp_map)

df['BsmtFin Type 1'] = df['BsmtFin Type 1'].map(bsmt_fin_map)
df['BsmtFin Type 2'] = df['BsmtFin Type 2'].map(bsmt_fin_map)

df['Garage Finish'] = df['Garage Finish'].map(garage_finish_map)

df['Paved Drive'] = df['Paved Drive'].map(drive_map)

df['Land Slope'] = df['Land Slope'].map(slope_map)

df['Lot Shape'] = df['Lot Shape'].map(shape_map)
```

```python
fill_zero_cols = [
    'Bsmt Qual', 'Bsmt Cond', 'Fireplace Qu',
    'Garage Qual', 'Garage Cond', 'Pool QC',
    'Garage Cars', 'Garage Area',
    'Bsmt Full Bath', 'Bsmt Half Bath'
]
df[fill_zero_cols] = df[fill_zero_cols].fillna(0).astype(int)
# so that we get the dtype as int instead of float with thesse there are still nan which makes them float
```

```python
# for central air the model doesnot understand y and n which mean yes and no so changing it to  1 and 0
df['Central Air'] = df['Central Air'].map({'Y': 1, 'N': 0})
```

### fixing typos in the dataset

```python
df.loc[df['Garage Yr Blt'] == 2207, 'Garage Yr Blt'] = 2007
# the column has year 2207 which is obviously a typo so changing that
```

### removing columns that could make the model pick wrong things

```python
# here in this dataset have 5 rows where even big house are sold for very little price not because of the market 
# but because they are sold before completion and some which were sold hurriedly this obviously doesn't reflecty the current 
# market so we removed cuz the model might pick wrong things


df = df[~(
    (df['Gr Liv Area'] > 4000) &
    (df['Sale Condition'].isin(['Partial', 'Abnormal']))
)]
```

```python
df.shape
```

### log transformation and one hot encoding

```python
# without log tranfrom the model could think since the cheap house errors are small compared to the big house error
# which is totally wrong since the big house is pricier compared to small housde the model could think only the expensive house
# as anomalies

df['SalePrice'] = np.log1p(df['SalePrice'])
```

```python
df.dtypes
```

```python
ohe_cols = [col for col in df.columns if df[col].dtype == 'object']
ohe_cols += ['MS SubClass', 'Mo Sold']
df = pd.get_dummies(df, columns=ohe_cols, drop_first=True, dtype=int)
```

```python
print("Object cols left:", df.select_dtypes(include='object').columns.tolist())
print("Total nulls:", df.isnull().sum().sum())
print("Shape:", df.shape)
```

```python
df.to_csv('ames_cleaned.csv', index=False)
# extarct the cleaned dataset
```
